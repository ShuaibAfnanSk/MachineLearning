```python
df['Source'] = df['Source'].str.replace('Original','Source : Original ')
df['Source'] = df['Source'].str.replace('Manga','Source : Manga ')
df['Source'] = df['Source'].str.replace('Light novel','Source : Light novel ')
df['Source'] = df['Source'].str.replace('Game','Source : Game ')
df['Source'] = df['Source'].str.replace('Visual novel','Source : Visual novel ')
df['Source'] = df['Source'].str.replace('4-koma manga','Source : 4-koma manga ')
df['Source'] = df['Source'].str.replace('Unknown','Source : Unknown ')
df['Source'] = df['Source'].str.replace('Picture book','Source : Picture book ')
df['Source'] = df['Source'].str.replace('Web manga','Source : Web manga ')
df['Source'] = df['Source'].str.replace('Music','Source : Music ')
df['Source'] = df['Source'].str.replace('Radio','Source : Radio ')
df['Source'] = df['Source'].str.replace('Book','Source : Book ')
df['Source'] = df['Source'].str.replace('Card game','Source : Card game ')
df['Source'] = df['Source'].str.replace('Digital manga','Source : Digital manga ')
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(stemmer.stem(i))
    return " ".join(y)
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = vectorizer.fit_transform(data['tags']).toarray()
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[2])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
similarity.shape
```

```python
def recommend(anime):
    anime_index = data[data['Name'] == anime].index[0]
    distances = similarity[anime_index]
    anime_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in anime_list:
        print(data.iloc[i[0]].Name)
        print(data.iloc[i[0]].Score)
        print(data.iloc[i[0]].Type)
```

```python
recommend('Gintama')
```
