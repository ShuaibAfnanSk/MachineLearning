```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('anime.csv')
df2 = pd.read_csv('anime_with_synopsis.csv')
df = df1.merge(df2,on='MAL_ID')
df
```

```python
df.columns
```

```python
trending = df[['MAL_ID','Name_x','Score_x','Type','Popularity','Favorites','Watching','Completed','Dropped','On-Hold']]
```

```python
trending['Trend_rate'] = trending['Watching'] + trending['Completed'] / trending['Dropped'] + trending['On-Hold']
trending['Trend_rate'] = trending['Trend_rate'].replace(np.inf,np.nan)
trending.dropna(inplace=True)
```

```python
trending = trending.sort_values(ascending=False,by='Trend_rate').head(25)
```

```python
import pickle
pickle.dump(trending,open('trend.pickle','wb'))
```
