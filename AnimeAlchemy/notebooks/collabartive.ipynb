```python
import pandas as pd
import numpy as np
```

```python
anime = pd.read_csv('anime.csv')
```

```python
anime.columns
```

```python
anime['anime_id'] = anime['MAL_ID']
```

```python
anime = anime[['anime_id','Name','Score','Genres','Type']]
```

```python
anime
```

```python
rating = pd.read_csv('rating_complete.csv')
```

```python
df = rating.merge(anime,on='anime_id')
```

```python
df
```

```python
rated_anime = df.groupby('anime_id').count()['rating'] >= 10000
```

```python
top_rated = rated_anime[rated_anime].index
```

```python
dataset = df[df['anime_id'].isin(top_rated)]
```

```python
np.where(dataset['Name']=='Gintama')
```

```python
raters = dataset.groupby('user_id').count()['rating'] >= 500
```

```python
top_raters = raters[raters].index
```

```python
data = dataset[dataset['anime_id'].isin(top_rated)]
```

```python
np.where(data['Name']=="One Punch Man")
```

```python
pivot = data.pivot_table(index='Name',columns='user_id',values='rating')
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pivot)
```

```python
np.where(pivot.index=='Gintama')[0][0]
```

```python
def recommend(anime_name):
    index = np.where(pivot.index==anime_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = anime[anime['Name'] == pivot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Name')['Name'].values))
        item.extend(list(temp_df.drop_duplicates('Name')['Genres'].values))
        item.extend(list(temp_df.drop_duplicates('Name')['Score'].values))
        item.extend(list(temp_df.drop_duplicates('Name')['Type'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('One Punch Man')
```
